```python
import bcrypt
import re

def is_strong_password(password: str) -> bool:
    if len(password) < 8:
        return False
    if not re.search(r"[A-Z]", password):
        return False
    if not re.search(r"[a-z]", password):
        return False
    if not re.search(r"[0-9]", password):
        return False
    if not re.search(r"[!@#$%^&*()_+=\-{}[\]:;\"'<>,.?/]", password):
        return False
    return True

def hash_password(password: str) -> bytes:
    password_bytes = password.encode("utf-8")
    salt = bcrypt.gensalt(rounds=12)
    return bcrypt.hashpw(password_bytes, salt)

def verify_password(password: str, hashed_password: bytes) -> bool:
    return bcrypt.checkpw(password.encode("utf-8"), hashed_password)

def main():
    print("USER REGISTRATION")
    while True:
        password = input("Create a strong password: ")
        if is_strong_password(password):
            break
        else:
            print("Weak password!")
            print("Must be 8+ chars with uppercase, lowercase, number & symbol\n")
    stored_hash = hash_password(password)
    print("\nPassword stored securely!\n")
    print("USER LOGIN")
    login_password = input("Enter password to login: ")
    if verify_password(login_password, stored_hash):
        print("\nAuthentication Successful!")
    else:
        print("\nAuthentication Failed! Wrong password.")

if __name__ == "__main__":
    main()
```
